# Median havesine distance between pickup and dropoff location 

```python
import math
```

```python
# creating function for haversine distance 
def haversine_distance(latitude1, longitude1, latitude2, longitude2):
    #readius of earth in kms
    radius = 6371.0

    dlat = np.radians(latitude2 - latitude1)
    

    dlon = np.radians(longitude2 - longitude1)
    
    # Haversine formula
    a = np.sin(dlat / 2)**2 + np.cos(np.radians(latitude1)) * np.cos(np.radians(latitude2)) * np.sin(dlon / 2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))

    distance = radius * c
    
    return distance
```

```python
df['haversine_distance'] = df.apply(lambda row: haversine_distance(row['pickup_latitude'], row['pickup_longitude'], row['dropoff_latitude'], row['dropoff_longitude']), axis=1)
```

# Apply a Machine Learning Algorithm to predict the fare amount given following input features: passenger_count, distance and ride_week_day
# Perform a 70-30 split of data
# Which algorithm gives the least adjusted R square value

```python
# importing required methods,clases algorithms and metrics

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score
```

```python
df
```

```python
X = df[['passenger_count', 'haversine_distance', 'day_of_week']]
y = df['fare_amount']
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train_cat = X_train.select_dtypes(include=['object'])

X_train_cat.head()
```

```python
X_train_num = X_train.select_dtypes(include=['int64', 'float64'])

X_train_num.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_num_rescaled = pd.DataFrame(scaler.fit_transform(X_train_num), 
                                    columns = X_train_num.columns, 
                                    index = X_train_num.index)

X_train_num_rescaled.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder_ = OneHotEncoder(drop='first', sparse=False)

X_train_cat_ohe = pd.DataFrame(encoder_.fit_transform(X_train_cat), 
                               columns=encoder_.get_feature_names_out(X_train_cat.columns), 
                               index = X_train_cat.index)

X_train_cat_ohe.head()
```

```python
X_test_cat = X_test.select_dtypes(include=['object'])

X_test_cat.head()
```

```python
X_test_num = X_test.select_dtypes(include=['int64', 'float64'])

X_test_num.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_test_num_rescaled = pd.DataFrame(scaler.fit_transform(X_test_num), 
                                    columns = X_test_num.columns, 
                                    index = X_test_num.index)

X_test_num_rescaled.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder_ = OneHotEncoder(drop='first', sparse=False)

X_test_cat_ohe = pd.DataFrame(encoder_.fit_transform(X_test_cat), 
                               columns=encoder_.get_feature_names_out(X_test_cat.columns), 
                               index = X_test_cat.index)

X_test_cat_ohe.head()
```

```python
X_train_transformed = pd.concat([X_train_num_rescaled, X_train_cat_ohe], axis=1)

X_train_transformed.head()
```

```python
X_test_transformed = pd.concat([X_test_num_rescaled, X_test_cat_ohe], axis=1)

X_test_transformed.head()
```

```python
models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree Regression': DecisionTreeRegressor(),
    'Random Forest Regression': RandomForestRegressor(),
    'K-Nearest Neighbors Regression': KNeighborsRegressor()
}
```

```python
adjusted_r_squared_values = {}

for model_name, model in models.items():
    model.fit(X_train_transformed, y_train)
    y_pred = model.predict(X_test_transformed)
    
    # Calculate the R-squared value
    r_squared = r2_score(y_test, y_pred)
    
    # Calculate the adjusted R-squared value
    n = X_test_transformed.shape[0]  # Number of samples
    p = X_test_transformed.shape[1]  # Number of predictors (features)
    adjusted_r_squared = 1 - ((1 - r_squared) * (n - 1) / (n - p - 1))
    
    adjusted_r_squared_values[model_name] = adjusted_r_squared
```

```python
min_adj_r_squared_model = min(adjusted_r_squared_values, key=adjusted_r_squared_values.get)
min_adj_r_squared_value = adjusted_r_squared_values[min_adj_r_squared_model]
```

```python
print("Adjusted R-squared values for different models on the test data:")
```

```python
for model_name, adj_r_squared in adjusted_r_squared_values.items():
    print(model_name,":", adj_r_squared)
```

```python
print("The model with the least adjusted R-squared value on the test data is ", min_adj_r_squared_model," with a value of ", min_adj_r_squared_value,":")
```
